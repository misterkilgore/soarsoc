### 🧩 1. Struttura generale dei dati

Ogni feature ha una **parte principale** (es. `radius`, `smoothness`, ecc.) e un **prefisso**:

* `mean` → media delle misure su tutte le cellule del campione;
* `error` → deviazione standard o errore medio (variazione tra le cellule);
* `worst` → il valore massimo o più anomalo (ad esempio, la cellula più grande o irregolare).

---

### 🔬 2. Significato delle caratteristiche principali

| Feature               | Significato (intuitivo)                        | Interpretazione biologica                   |
| --------------------- | ---------------------------------------------- | ------------------------------------------- |
| **radius**            | Distanza media dal centro del nucleo al bordo  | Dimensione generale del nucleo              |
| **texture**           | Variazione d’intensità dei pixel nell’immagine | Irregolarità della struttura del nucleo     |
| **perimeter**         | Lunghezza del contorno del nucleo              | Quanto è grande o complesso il bordo        |
| **area**              | Area racchiusa dal contorno                    | Grandezza totale del nucleo                 |
| **smoothness**        | Variazione locale del raggio                   | Quanto è “liscio” o “frastagliato” il bordo |
| **compactness**       | (Perimeter² / Area) − 1                        | Quanto il nucleo è denso o compatto         |
| **concavity**         | Grado di concavità delle rientranze sul bordo  | Deformazioni tipiche di cellule tumorali    |
| **concave points**    | Numero di punti concavi sul contorno           | Quante rientranze sono presenti             |
| **symmetry**          | Misura della simmetria del nucleo              | Tumori spesso mostrano nuclei asimmetrici   |
| **fractal dimension** | Indicatore di complessità del contorno         | Quanto il bordo è “frattale” o irregolare   |

---

### 📊 3. Esempio di lettura

* `mean radius`: dimensione media dei nuclei nel campione
* `radius error`: variabilità della dimensione tra i nuclei
* `worst radius`: dimensione del nucleo più grande (più “estremo”)






```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

import pandas as pd

# Carico il dataset
data = load_breast_cancer()
```

```python
X = data.data
y = data.target

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Modello
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Predizione
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc
```
